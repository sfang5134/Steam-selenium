```python
#importing the webdriver library from selenium
from selenium import webdriver
#importing the Options libary from the chrome options
from selenium.webdriver.chrome.options import Options
#the BY keyboard tells specifies what parameter to limit to search to
#E.g. ID, CLASS_NAME, etc.
from selenium.webdriver.common.by import By

#wait and EC allows the program to wait while the webpage loads
from selenium.webdriver.support.ui import WebDriverWait as wait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# Create a Chrome web driver and setting the options
chrome_options = webdriver.ChromeOptions()

# Keep the browser window open until the user closes it
chrome_options.add_experimental_option("detach", True)

# Instantiating the driver
driver = webdriver.Chrome(options=chrome_options)

# Setting the url for Steam store
url = "https://store.steampowered.com/search/?filter=topsellers"
driver.get(url)

# Keywords for the video games
keywords = ["action", "adventure", "strategy", "rpg", "sports", "simulation", "indie", "horror", "puzzle", "multiplayer"]

# Lists to store data
product_name_list = []
product_price_list = []
link_list = []

for keyword in keywords:
    # Selecting the search box by using its ID
    search = driver.find_element(By.ID, 'store_nav_search_term')

    # Using send keys to enter text into the form field using .sendkeys()
    search.send_keys(keyword)

    # Waiting for the search results to load
    time.sleep(2)

    # Clicking on the search button
    search_button = driver.find_element(By.CLASS_NAME, 'searchbox_button')
    search_button.click()

    # Collecting a list of all the divs which contain the search results
    scraped_items = wait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, '//a[@class="search_result_row ds_collapse_flag app_impression_tracked"]')))

    # Iterating through the list of divs to find specifics of each item in the search result
    for item in scraped_items:
        # Find name
        product_name = item.find_element(By.CLASS_NAME, 'title').text

        # Find price
        product_price = item.find_element(By.CLASS_NAME, 'search_price').text.strip()

        # Find link
        link = item.get_attribute('href')

        # Append scraped information to respective lists
        product_name_list.append(product_name)
        product_price_list.append(product_price)
        link_list.append(link)

        print(product_name)
        print(product_price)
        print(link)
        print('\n')

# Creating a DataFrame to store the data
df = pd.DataFrame({
                    'Product Name': product_name_list,
                    'Product Price': product_price_list,
                    'Link': link_list
})

# Exporting the data to a CSV file
df.to_csv('video_games.csv', index=False)

# Close the browser
driver.quit()
```
